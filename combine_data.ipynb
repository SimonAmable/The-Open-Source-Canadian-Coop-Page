```python
from pymongo import MongoClient
```

# Setup Variables

```python
mongo = MongoClient('mongodb://localhost:27017/')
db = mongo['my_database']
glassdoor_collection = db['opportunities_glassdoor']
levels_collection = db['opportunities_levels_fyi']
linkedin_collection = db['opportunities_linkedIn']
all_jobs_collection = db['all_jobs']

```

```python
# Combine Data
```

```python
#----- Combine Data from all collections into "all_jobs" -----
# Add documents from one collection to another

linkedin_documents = list(linkedin_collection.find())
all_jobs_collection.insert_many(linkedin_documents)

glassdoor_documents = list(glassdoor_collection.find())
all_jobs_collection.insert_many(glassdoor_documents)

levels_documents = list(levels_collection.find())
all_jobs_collection.insert_many(levels_documents)
```
